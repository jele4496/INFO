#### Class initialization __ init __ and methods

```python
# make a class with a function that intializes a name and age as arguments
# make a method that prints out the name and age of the 

class Person():
    #sets arguments and sets the scope
    def __init__(self, name, age):
        self.name = name
        self.age = age

    #method
    def qualities(self):
        print('name: ', self.name, ', age: ', str(self.age))

```

```python
#create an object/ class instance

student1 = Person('Paul', 20)
```

```python
# what is this object?
student1
```

```python
#get the name variable within the object

student1.name
```

```python
#get the age variable within the object
student1.age
```

```python
# call the method within the class
student1.qualities()
```

```python
# make a class with name, age, and with a keyword argument of occupation with 'student 
class Student():
    def __init__(self, name, age, occupation = 'student'):
        self.name = name
        self.age = age
        self.occupation = occupation

# create a method with to print all of the argumetns
    def printMethod(self):
        print(self.name, self.age, self.occupation)
```

```python
# create an instance
george = Student('George', 23, 'teacher')

george.printMethod()
```

```python
# make a class with name, age, and with a keyword argument of occupation with 'student 

class Student_advanced():
    def __init__(self, name, age, occupation = 'student'):
        self.name = name
        self.age = age
        self.occupation = occupation

# create a method with to print all of the argumetns
    def printMethod(self):
        print(self.name, self.age, self.occupation)

# create a method that takes in a list of grades and prints the sum of those grades
    def sumOfGrades(self, listOfGrades):
        self.listOfGrades = listOfGrades #scope of the method but a part of the class once initialized
        print(self.name, 'total grade is ', sum(self.listOfGrades))

# create a method that prints out  the list of grades
    def printGrades(self):
        print(self.listOfGrades)
```

```python
kerry = Student_advanced('kerry', 15)

kerry.sumOfGrades([99,90,98,87,60])
```

```python
kerry.listOfGrades
```

```python
luna = Student_advanced('luna', 4, 'dog')

# luna.printMethod()
luna.listOfGrades
```

```python
luna.sumOfGrades([20,38,80])

print(luna.listOfGrades)
```

```python
# what is the type of this instance?
person1 = Student_advanced('florence', 80)
person1.printMethod()

print(type(person1))
```

```python
# create a class that assigns two arguments and 
# has a method that prints the arguments
# has a method that has a list of tasks

class DogWash():
    def __init__(self, dog, age):
        self.dog = dog
        self.age = age

    def dogInfo(self):
        print('name = ', self.dog,'age = ', self.age)

    def getTasks(self, taskLyst): #ONLY scoped for the method
        return(taskLyst)
```

```python
# assign an instance with a name and an age
inigo_montoya = DogWash('inigo montoya', 10)

inigo_montoya.dogInfo()
```

```python
# define the list 
day1_Tasks = inigo_montoya.getTasks(['brush', 'clean', 'cut'])
day1_Tasks
```

```python
day2_Tasks = inigo_montoya.getTasks(['cuddle', 'fetch'])
day2_Tasks
```
