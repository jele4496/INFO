A function is a block of code which only runs when it is called.
You can pass data, known as parameters, into a function.
A function can return data as a result.

In Python a function is defined using the def keyword.

Information can be passed into functions as arguments.
Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.
*The terms parameter and argument can be used for the same thing: information that are passed into a function.*
From a function's perspective:

----A parameter is the variable listed inside the parentheses in the function definition.

----An argument is the value that is sent to the function when it is called.

### Scopes

a funciton has its own scope. Things that are mutated within the funciton aren't necessarily changed outside of the function. This means that you can call and create a variable outside of the function and reference the a variable by the same name. However, this doesn't overwright the original variable because the second one is only referencable and called within the function

```python
# assign a variable
myVariable = 10

# make a function that has the variable as an argument
# have it return the mutated variable
def update_myVariable(myVariable):
    myVariable += 5
    return myVariable


#print the output of the funciton and the original variable
print('outvariable--', myVariable)
print('invariable --', update_myVariable(900))
print('callInFunction --', update_myVariable(myVariable))
```

```python
# make a variable
cool = 52

# make a function that prints the variable with 5 added to it
def newFunction(cool):
    print(cool + 5)


# call the funciton and print the original variable
print('outvariable--', cool)
print('invariable --', newFunction(10))
# newFunction(10)
```

### Arbitrary Arguments, *args


This way the function will receive a tuple of arguments, and can access the items accordingly:

```python
# make function that can take in as many arguments as it needs
# print the 2nd item in the arguments

def arbArgFunction(*lyst):
    print(lyst)
    print(lyst[1])
```

```python
arbArgFunction(2,3,4,5,6,7,7,7)
```

```python
# make function that can take in as many arguments as it needs
# make the arguments a list of names
# print 'The youngest child is'... and then add the first name in the list of names

def arbArgFunction2(*kid):
    print(kid)
    print('The youngest child is ', kid[0])
```

## Excercise 2

You roll two six-sided dice, each with faces containing one, two, three, four, five and six spots, respectively. 

When the dice come to rest, the sum of the spots on the two upward faces is calculated. 
- If the sum is 7 or 11 on the first roll, you win. 
- If the sum is 2, 3 or 12 on the first roll (called “craps”), you lose (i.e., the “house” wins). 
- If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your “point.” 
    - To win, you must continue rolling the dice until you “make your point” (i.e., roll that same point value). You lose by rolling a 7 before making your point.

```python
def rollDice():
    dice1 = random.randrange(1, 7)
    dice2 = random.randrange(1, 7)
    return dice1, dice2
```

```python
d1, d2 = rollDice()
```

```python
d1, d2 = rollDice()
tot = d1 + d2
print(tot)

pointslyst = [4, 5, 6, 8, 9, 10]

if (tot == 7  or tot == 11):
    print('you win!')
if (tot == 2 or tot == 3 or tot == 12):
    print('craps! you lose')
if tot in pointslyst:
    print(tot, 'is your point')

    win = False

    while(win == False): 
        d3,d4 = rollDice()
        subtot = d3 + d4
        print('dice total = ', subtot)

        if subtot == tot:
            win = True
            print('you win')

        elif subtot == 7:
            print('you lose')
            break
        



```
